```python
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get("https://www.youtube.com/")
response.status_code
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
soup.title
```

```python
#code to extract all the links from youtube homepage
link_list=[]
for link in soup.find_all('a'):
    if str(link.get('href')).startswith("/watch?v=") and "https://www.youtube.com"+str(link.get('href')) not in link_list:
        link_list.append("https://www.youtube.com"+link.get('href'))

  
   
```

```python
#code to iterate over each link extracted above and extract likes,dislikes,views,titles,channels,thumnail links   
like_list=[]
view_list=[]
dislike_list=[]
title_list=[]
channel_list=[]
thumb_list=[]

for link in link_list[:20]:
    response2=requests.get(link)
    soup2=BeautifulSoup(response2.content,"html.parser")
    for likes in soup2.find_all('button'):
        if str(likes.get('aria-label')).startswith("like"):
            
            like_list.append(str(likes.get('aria-label')).split()[5])
            break
                             
    for dislikes in soup2.find_all('button'):
        if str(dislikes.get('aria-label')).startswith("dislike"):
            dislike_list.append(str(dislikes.get('aria-label')).split()[5])
            break
                                
    for views in soup2.find_all('div',{'class':'watch-view-count'}):
        view_list.append(views.string)
        
    for link in soup2.find_all('title'):
        if link.string not in title_list:
            title_list.append(link.string)
            
    string=''
    for channel in soup2.find_all('a',{'class':'yt-uix-sessionlink spf-link'}):
        
        if str(channel.get('href')).startswith("/channel"):
            string=string+channel.text
            string=string+" "
            
    channel_list.append(string)
    
    for i in soup2.find_all('link',{'itemprop':'thumbnailUrl'}):
        thumb_list.append(i.get("href"))
            
    






```

```python
#code to merge all the lists into one tuple and create a dataframe from it 
import pandas as pd
data_tuples= list(zip(title_list,link_list[:20],like_list,dislike_list,view_list,channel_list,thumb_list))
df=pd.DataFrame(data_tuples, columns=['title','link','like','dislike','view','channel','thumbnail'])
df




```

```python
#code to write data frame to excel file 
df.to_excel("youtube_data.xlsx")
```
